### Vectorization & Broadcasting

<b>1. Vectorization</b>: It is the process of applying operations to entire arrays or sequences of data at once, rather than iterating through individual elements using explicit loops. 

```python
# Vectorization & Broadcasting

#Vectorization
arr = np.array([1, 2, 3, 4, 5])

sq_arr = arr ** 2       #square of all nums
print(sq_arr)

arr2 = np.array([6, 7, 8, 9, 10])
print(arr + arr2)      #sum of 2 arrays

#Broadcasting
arr_mul10 = arr * 10    #multiply by 10 to all nums
print(arr_mul10)

arr1D = np.array([1, 2, 3])
arr2D = np.array([[1, 2, 3], [4, 5, 6]])
print(arr1D + arr2D)
```

Broadcasting: (row, column) = m x n = shape

now to broadcast two arrays or perform operations on them, they need to be compatible. 

How to check compatibility? 
- Take 2 arrays:
(r1, c1) and (r2, c2) 
- You start checking/comparing from right value(columnns)
  i.e. compare c1, c2 first

  2 rules:
  - Either they c1 = c2
  - Or c1 = 1 or c2 =1

- If the above condition is valid then do the same for r1, r2.

If both conditions satisfy, then the shapes are compatible.
