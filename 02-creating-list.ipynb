# Creating NumPy Arrays

Covers two ways to create NumPy arrays:
1. **From existing Python lists/sequences** — `np.array()`, mixed-type coercion, 2-D matrices.
2. **From scratch** — `zeros`, `ones`, `full`, `eye` (identity), `arange` (range-like), `linspace` (evenly spaced).

## Creating np arrays using lists

```python
import numpy as np
```

```python
arr = np.array([1,2,3,4,5])

# Mixed types: NumPy upcasts the whole array to the most general type (string here)
arr2 = np.array([1, 2, 3, 4, "prime"])
print(arr2, type(arr2))

# 2D Arrays — nested lists become a matrix; shape is (rows, cols)
arr3 = np.array([[1, 2, 3], [4, 5, 6]])
print(arr3, arr3.shape)
```

### This is the N-Dimensional array data structure of NumPy

```python
print(arr)
print(type(arr))
```

## Other ways of Creating Arrays are:

```python
# Creating NumPy Arrays — from scratch (no existing data needed)
arr1 = np.zeros((3, 4))        # 3×4 matrix filled with 0.0
print(arr1, arr1.shape)

arr2 = np.ones((3, 3))         # 3×3 matrix filled with 1.0
print(arr2, arr2.shape)

arr3 = np.full((2, 3), 5)      # 2×3 matrix filled with the constant 5
print(arr3, arr3.shape)

arr4 = np.eye(3)               # 3×3 identity matrix (1s on diagonal)
print(arr4)

arr5 = np.arange(1, 20, 2)    # integers 1 to 19, step 2 (like Python range)
print(arr5)

arr6 = np.linspace(0, 10, 5)  # 5 evenly-spaced values between 0 and 10 (inclusive)
print(arr6)
```
