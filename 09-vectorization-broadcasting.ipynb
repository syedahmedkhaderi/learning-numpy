# Vectorization & Broadcasting

- **Vectorization**: applying operations to entire arrays at once — no Python `for` loops needed, much faster.
- **Broadcasting**: NumPy automatically "stretches" smaller arrays to match larger ones during operations, as long as their shapes are compatible. Compatibility rule (checked right-to-left): dimensions must be equal, or one of them must be 1.

***

### Vectorization & Broadcasting

<b>1. Vectorization</b>: It is the process of applying operations to entire arrays or sequences of data at once, rather than iterating through individual elements using explicit loops. 

```python
import numpy as np

# Vectorization — element-wise operations applied to the whole array without a loop
arr = np.array([1, 2, 3, 4, 5])

sq_arr = arr ** 2          # squares each element: [1, 4, 9, 16, 25]
print(sq_arr)

arr2 = np.array([6, 7, 8, 9, 10])
print(arr + arr2)          # element-wise addition of two same-shape arrays: [7, 9, 11, 13, 15]

# Broadcasting — scalar is "broadcast" to match the array's shape automatically
arr_mul10 = arr * 10       # every element multiplied by 10
print(arr_mul10)

# Broadcasting between 1-D and 2-D: the 1-D row is added to each row of the 2-D array
arr1D = np.array([1, 2, 3])
arr2D = np.array([[1, 2, 3], [4, 5, 6]])
print(arr1D + arr2D)  # arr1D is broadcast across both rows of arr2D
```

Broadcasting: (row, column) = m x n = shape

now to broadcast two arrays or perform operations on them, they need to be compatible. 

How to check compatibility? 
- Take 2 arrays:
(r1, c1) and (r2, c2) 
- You start checking/comparing from right value(columnns)
  i.e. compare c1, c2 first

  2 rules:
  - Either they c1 = c2
  - Or c1 = 1 or c2 =1

- If the above condition is valid then do the same for r1, r2.

If both conditions satisfy, then the shapes are compatible.
