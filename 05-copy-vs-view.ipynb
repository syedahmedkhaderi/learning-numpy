# Copy vs View in NumPy

A critical NumPy gotcha: slicing a NumPy array returns a **view** (shared memory), not a copy like Python lists do.  
- Modifying a view changes the original array.  
- Use `.copy()` to get an independent copy that won't affect the source.

```python
import numpy as np
```

```python
# Python list slice → independent copy; modifying it does NOT change the original
print("In Python: ")
py_list = [1, 2, 3, 4, 5]
copy_list = py_list[1:4]  # new list [2, 3, 4]
copy_list[1] = 333

print(copy_list)   # [2, 333, 4]
print(py_list)     # unchanged: [1, 2, 3, 4, 5]

# NumPy slice → view sharing the same memory; modifying it DOES change the original
print("\n Now For NumPy: ")
np_arr = np.array([1, 2, 3, 4, 5])
view_arr = np_arr[1:4]   # view into positions 1, 2, 3
view_arr[1] = 333        # writes through to np_arr[2]

print(view_arr)  # [  2 333   4]
print(np_arr)    # also changed: [  1   2 333   4   5]

# Use .copy() to get a truly independent slice
print("\n Now creating a copy: ")
copy_arr = np_arr[1:4].copy()  # detaches from np_arr
copy_arr[2] = 444
print(copy_arr)  # [  2 333 444]
print(np_arr)    # unchanged
```
