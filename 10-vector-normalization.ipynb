# Vector Normalization (Z-score / Standard Normalization)

A practical application of broadcasting in ML preprocessing. **Z-score normalization** transforms data so each feature has mean = 0 and standard deviation = 1:

```
x_normalized = (x − μ) / σ
```

This is essential before algorithms like gradient descent or PCA that are sensitive to feature scale.

A quite common example of broadcasting in Vector Normalization. This is very common in machine learning and data preprocessing.


Let’s take an example of Standard Vector Normalization i.e. transforming an array such that it has:


- mean = 0
- standard deviation = 1


For each element xi in vector, Where:
xi normalized = (xi− μ)/σ
- μ = mean of the vector (or column)
- σ = standard deviation

````python
import numpy as np
````

````python
# Vector Normalization — Z-score: (x - mean) / std_dev
# Result: mean ≈ 0, std ≈ 1 — broadcasting handles subtraction/division across the array

arr = np.array([[1, 2], [3, 4]])
mean = np.mean(arr)       # overall mean of all elements
std_dev = np.std(arr)     # overall std dev

print((arr - mean) / std_dev)  # each element normalised relative to the whole array

# Column-wise normalization — normalise each feature (column) independently
# This is standard practice in ML: each feature gets its own mean/std
arr = np.array([[1, 2], [3, 4], [5, 6]])
mean = np.mean(arr, axis=0)     # mean of each column (shape: (2,))
std_dev = np.std(arr, axis=0)   # std of each column (shape: (2,))
print((arr - mean) / std_dev)   # broadcasting subtracts/divides column-wise
````
