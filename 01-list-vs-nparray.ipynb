# NumPy Arrays vs Python Lists

Benchmarks speed and memory of Python lists versus NumPy arrays on 10 million elements. Key takeaways:
- NumPy arrays are **~24× faster** for element-wise operations due to vectorisation (no Python loop overhead).
- NumPy arrays use **~10 000× less memory** because elements are stored as homogeneous C-types, not boxed Python objects.

### Python lists vs Numpy Arrays

```python
import numpy as np
import time

size = 10_000_000  # 10 million elements — large enough to make timing differences visible

# --- Python List benchmark ---
python_list = list(range(size))

start = time.time()
list_squared = [x**2 for x in python_list]  # list comprehension iterates in Python bytecode
end = time.time()
print("Python list time:", end - start, "seconds")

# --- NumPy Array benchmark ---
numpy_array = np.array(python_list)
start = time.time()
array_squared = numpy_array ** 2  # vectorised: delegated to compiled C code, no Python loop
end = time.time()
print("NumPy array time:", end - start, "seconds")
```

### Comparing Memory

```python
# Memory usage comparison
# sys.getsizeof returns bytes per Python list object (each element is a pointer to a boxed int)
# numpy.nbytes is the raw byte footprint of the actual numeric data
import sys
print("Python list size:", sys.getsizeof(python_list) * len(python_list))
print("NumPy array size:", numpy_array.nbytes)
```
